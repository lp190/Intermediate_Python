## Creating a plot is one thing, now we learn customization

```python
# Basic scatter plot, log scales
plt.scatter(gp['gdp_cap'], gp['life_exp'])
plt.xscale('log')
```

```python
# Strings


plt.scatter(gp['gdp_cap'], gp['life_exp'])        
plt.xscale('log')
plt.xlabel('GDP per Capita in $')
plt.ylabel('Life Expectancy in years')
plt.title("World Development in 2007")  

########### Adapt ticks###########
##########################
# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']


plt.xticks(tick_val, tick_lab)
```

```python
# Change size of the dots as per the population
# Store population as a numpy array: np_pop
np_pop = np.array(gp['population'])


#Double np_pop
np_pop = np_pop * 0.000001

plt.scatter(gp['gdp_cap'], gp['life_exp'], s = np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])





```
